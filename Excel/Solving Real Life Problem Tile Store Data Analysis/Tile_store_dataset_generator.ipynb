```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
num_rows = 1000
num_customers = 250
start_date = datetime(2024, 1, 1)
end_date = datetime(2025, 12, 31)

tile_types = [
    "Ceramic Tile", "Porcelain Tile", "Marble Tile", "Granite Tile",
    "Mosaic Tile", "Vinyl Tile", "Terracotta Tile", "Glass Tile"
]

def random_date(start, end):
    delta = end - start
    rand_days = random.randrange(delta.days)
    return start + timedelta(days=rand_days)

data = {
    "order_id": [],
    "customer_id": [],
    "order_date": [],
    "product_type": [],
    "order_value": []
}

for i in range(num_rows):
    order_id = f"ORDER{str(i + 1).zfill(5)}"
    customer_id = f"CUST{str(random.randint(1, num_customers)).zfill(4)}"
    date = random_date(start_date, end_date)
    product = random.choice(tile_types)
    order_value = round(random.uniform(50, 10000), 2)

    data["order_id"].append(order_id)
    data["customer_id"].append(customer_id)
    data["order_date"].append(date)
    data["product_type"].append(product)
    data["order_value"].append(order_value)
```

```python
df = pd.DataFrame(data)

filename = "tile_store_orders.csv"
df.to_csv(
    filename,
    index=False,
    sep=';',
    decimal=',' # Use comma as decimal separator so excel works properly
)

print(f"Dataset created and saved as: {filename}")
print(df.head())
```
